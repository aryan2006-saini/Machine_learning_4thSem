<h1><b> 2. Codes + Outputs</b></h1>

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```


#ðŸ”¹ 1. Load the California Housing Dataset


```python
california = datasets.fetch_california_housing()

x = california.data
y = california.target

print(x)
```

```python
print(y)
```

```python
print("Feature Names:", california.feature_names)
print("Target Name:", california.target_names)
```

# ðŸ”¹ 2. Split the Dataset into Train & Test
    

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=11)
```

- 80% data â†’ training
- 20% data â†’ testing
- random_state=11 ensures reproducibility.

# ðŸ”¹ 3. Create and Train the Linear Regression Model

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

# ðŸ”¹ 4. Predict on Train & Test Data

```python
y_test_pred = model.predict(xtest)
y_train_pred = model.predict(xtrain)
```

# ðŸ”¹ 5. Evaluate the Model

```python
mse_test = mean_squared_error(ytest, y_test_pred)
mse_train = mean_squared_error(ytrain, y_train_pred)
r2_test = r2_score(ytest, y_test_pred)
r2_train = r2_score(ytrain, y_train_pred)

print("Intercept =", model.intercept_)
print("Coefficients =", model.coef_)
print("Train MSE =", mse_train)
print("Test MSE =", mse_test)
print("Train R2 =", r2_train)
print("Test R2 =", r2_test)
```

# ðŸ”¹ 6. Plot Actual vs Predicted (Training Data)

```python
plt.figure(figsize=(7, 6))
plt.scatter(ytrain, y_train_pred, alpha=0.5)
plt.plot([min(ytrain), max(ytrain)],[min(ytrain), max(ytrain)],'r--', linewidth=2)

plt.xlabel("Actual Values (y_train)")
plt.ylabel("Predicted Values (y_train_pred)")
plt.title("Actual vs Predicted (Training Set)")
plt.show()
```
